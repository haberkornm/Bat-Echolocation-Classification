# Bat Echolocation EDA

## Introduction to EDA

In this notebook, two bat echolocation libraries were examined, the first library was developed in Arizona and contained 5,849 calls.  The second library was developed in Mexico and contains 1,500 calls. The Arizona Library was developed through the use of Kaleidoscope Pro software to ID sonograms to bat species and then positively IDing species manually.  Software ID was then compared to the positive manual ID to determine accuracy of the computer software ID.  The Mexico Library was positively IDed by physically catching bats, identifying them, and then recording echolocation calls.  Ten features of the echolocation calls for both libraries were examined and anylized, and both libraries were compared to each other here for the purpose of developing a machine learning model to identify calls to species.

# Modifying Dataframe According to above findings

The Mexico and Arizona bat call libraries were only found to be compatible for the Fc, Dur, Fmin, and PkDur call features.  All other call features were found to either have slightly differnet definitions of the parameters of be measured on different scales.  Sc and S1 are both related to the slope of the call.  The Arizona library measures calls in octaves per millisecond while the Mexico library measures in kHz per millisecond.  Other features such as Fk and Tk have the same definition between the call libraries but differencer in where they exactly measure the k (knee of the call).  Fmax and BwDth also have the same definitions between the two libraries, however, the Mexico library only selected high quality search phase calls for their library.  The Arizona library was far less selective on the quality of calls and also included approach and feeding buzz calls.
